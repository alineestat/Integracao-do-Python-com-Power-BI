## 1. Configurações

* ### 1.1. Configurações necessárias Python

> #### Criar ambiente virtual do Python

Para funcionar a integração com Power BI é necessário utilizar a Versão 3.6 do Python. Necessário criar uma pasta com essa versão no ambiente virtual!

Manual do anaconda: https://docs.conda.io/projects/conda/en/4.6.0/_downloads/52a95608c49671267e40c689e0bc00ca/conda-
cheatsheet.pdf

O ambiente virtual será criado seguindo os passos abaixo. Com o prompt do Anaconda aberto, escrever a seguinte linha de código:

> Conda create --name envpowerbi python=3.6

Depois basta pressionar enter que a instalação será iniciada e depois de um tempo ele irá perguntar ao usuário se pretende prosseguir com a instalação, para isso basta escrever Y e pressionar enter novamente.

Feito isso o ambiente virtual já está instalado, no entanto ele vem vazio, ou seja, sem as bibliotecas que você já tinha ou qualquer outra.

Então será necessário instalar as bibliotecas dentro desse ambiente para que consiga utilizar o Python normalmente.

Lembrando que quando a instalação foi finalizada você ainda não está no ambiente virtual, pois vai ter o nome (base) ao lado esquerdo do caminho do seu computador.

Para acessar esse ambiente vamos utilizar o seguinte código:

> Conda activate envpowerbi

Assim que fizer isso, você vai notar que o nome (base) vai ser trocado para (envpowerbi) que é o nome do ambiente que criamos.

## 2. Código Python 

```python
import pandas as pd
from pandas_datareader import data as web
import matplotlib.pyplot as plt
import seaborn as sns
import ipeadatapy
```

```python
df = web.DataReader(f'^BVSP', data_source='yahoo', start=f'02-20-2020', end = '03-20-2020')
df = df.reset_index()
```

```python
display(df)
```

## 4. Visualização de gráficos do Python no Power BI

* IMPORTANTE: Para conseguir gerar um gráfico feito no Python dentro do Power BI é necessário antes ter a base de dados a ser utilizada para construção do gráfico dentro do Power BI. Para isso, pode-se exportar a base de dados do python para um arquivo excel ou rodar todo o código dentro do script PY do Power BI.

> Para criar um gráfico do Python dentro do Power BI, seguir os passos abaixo:
    
* Na guia Relatório ir em Visualizações -> Valores (Arrastar todas as colunas necesárias para construção do gráfico, formatar no tipo adequado) -> Clicar em PY (Visual Python) -> Abaixo da guia de Relatório vai aparecer o Script para adicionar o código do Python (Observa-se que o Power BI criou um dataframe com as variáveis selecionadas para fazer o gráfico. Adicionar o código do gráfico, mas com o nome do dataframe criado pelo Power BI, colocar todas as bibliotecas utilizadas).

```python
plt.figure(figsize=(15,5))
plt.plot(df["Date"], df["Adj Close"], "b--")
plt.tick_params(axis='x', labelrotation= 90)
plt.show()
```

Ao Colocar o código acima no Script do Power BI para gerar o gráfico, observa-se que o eixo x (com a data) está com a formatação diferente que do Python. Isso ocorre pois, apesar das bases do Power BI e do Python estarem com a coluna Date no formato de data, ambos possuem formatação padrão diferentes.
Para resolver o problema da formatação do eixo com a data é necessário deixar a data do Python no mesmo formato do Power BI.
Para resolver o problema da formatação da data basta usar o código abaixo:


```python
df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m-%dT%H:%M:%S')
```
